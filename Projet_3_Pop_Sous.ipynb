```python
# le premier je fais la liste pour m'aider filtrer d'annee que j'ai besoin dans csv files sousalimentation
# j'ai fais une condition - si dans le colonne Année il y a "2012-2014" on va ajouter dans nouvelle colonne  valeur "2013"
# mais si non il va ajouter "-"

LT_Annee_New = []

for i in np.arange(len(df_sousalimentation.index)):
        if df_sousalimentation.Année[i] == "2012-2014":
            LT_Annee_New.append("2013")
        else:
            LT_Annee_New.append("-")
            
# Transformer les listes en Dataframe Tmp, et ensuite lier avec dataframe d'origine

DF_TmpAnnee = pd.DataFrame({'Annee_Unique' : LT_Annee_New})
df_sousalimentation = pd.concat([df_sousalimentation,DF_TmpAnnee], axis=1)
df_sousalimentation['Annee_Unique'].head(5)
```

```python
# avec ces diffèrentes methodes on voit qui il y a des doublons dans "Zone" à cause de la Chine
# on voit que on dois conserve que le premier Chine

df_population = df_population[df_population.Zone.str.contains('Chine,') ==False]
df_population = df_population[df_population.Zone.str.contains('Chine -') ==False]
df_sousalimentation = df_sousalimentation[df_sousalimentation.Zone.str.contains('Chine,') ==False]
df_sousalimentation = df_sousalimentation[df_sousalimentation.Zone.str.contains('Chine -') ==False]

somme_df_population=df_population['Population'].sum()
print("Population mondial en 2013 :{} humains".format(somme_df_population))
```

```python
# On doit faire jointure entre pop2013_df et sousalim2013_df

pop_sous = pd.merge(pop2013_df,sousalim2013_df,on=["Code zone","Zone"], how="left")

pop_sous
```

```python
#Suprimer colonne Annee permently

pop_sous.drop(['Annee'], axis=1, inplace=True)
```

```python
#Calcul ratio _ percentage 
#j'ai choisi methode parce que La fonction iterrows () est responsable du passage en boucle de chaque ligne du DataFrame.

for ind,row in pop_sous.iterrows():
    pop_sous.loc[ind,"ratio_sous_personne"]=row['nb_sousalim']*100/row['Population']
    
pop_sous
```

```python
#Calcul ratio _ percentage sans iterows
pop_sous["ratio_sous_personne"] = round(pop_sous["nb_sousalim"]*100 / pop_sous["Population"],2)
pop_sous
```
